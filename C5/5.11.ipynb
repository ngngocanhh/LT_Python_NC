```python
import sqlite3

conn = sqlite3.connect('qly_nhan_vien.db')
cursor = conn.cursor()

# 1. Tạo bảng PHONG (Phòng ban)
cursor.execute('''CREATE TABLE IF NOT EXISTS PHONG (
    id INTEGER PRIMARY KEY,
    ten_phong TEXT NOT NULL)''')

# 2. Tạo bảng NHAN_VIEN (có khóa ngoại id_phong)
cursor.execute('''CREATE TABLE IF NOT EXISTS NHAN_VIEN (
    id INTEGER PRIMARY KEY,
    ho_ten TEXT NOT NULL,
    id_phong INTEGER,
    FOREIGN KEY (id_phong) REFERENCES PHONG(id))''')

# 3. Hàm thêm dữ liệu nhanh
def them_phong(id, ten):
    cursor.execute("INSERT INTO PHONG VALUES (?, ?)", (id, ten))

def them_nv(id, ten, id_phong):
    cursor.execute("INSERT INTO NHAN_VIEN VALUES (?, ?, ?)", (id, ten, id_phong))

# Lưu và đóng
conn.commit()
print("Đã tạo CSDL và bảng thành công.")
```
