```python
import sqlite3

conn = sqlite3.connect('product.db')
cursor = conn.cursor()

# Tạo bảng
cursor.execute('''CREATE TABLE IF NOT EXISTS product (
    Id INTEGER PRIMARY KEY, Name TEXT NOT NULL, Price REAL NOT NULL, Amount INTEGER NOT NULL)''')

# Các hàm chức năng
def xem(): 
    for r in cursor.execute("SELECT * FROM product"): print(r)

def them(n, p, a): 
    cursor.execute("INSERT INTO product (Name, Price, Amount) VALUES (?,?,?)", (n, p, a))

def tim(n): 
    for r in cursor.execute("SELECT * FROM product WHERE Name LIKE ?", (f'%{n}%',)): print(r)

def sua(id, p, a): 
    cursor.execute("UPDATE product SET Price=?, Amount=? WHERE Id=?", (p, a, id))

def xoa(id): 
    cursor.execute("DELETE FROM product WHERE Id=?", (id,))

# Lưu thay đổi
conn.commit()
```
