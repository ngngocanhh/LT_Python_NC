```python
data = [
    {"name": "An", "unit": "A1"},
    {"name": "Binh", "unit": "A1"},
    {"name": "Chi", "unit": "A2"},
    {"name": "Dung", "unit": "A3"},
    {"name": "Ha", "unit": "A3"},
    {"name": "Lan", "unit": "A4"}
]

print("Kết quả thực hiện chương trình:")
print("Tên công ty: Công ty TNHH Đất Việt")
print("Địa chỉ: abc Giải Phóng - Hà Nội.")
print("*-Thông kê nhân viên theo đơn vị..")

tong_nv = len(data)

thong_ke = {}

for kq in data:
    don_vi = kq["unit"]
    if don_vi in thong_ke:
        thong_ke[don_vi] += 1
    else:
        thong_ke[don_vi] = 1

stt = 1
for don_vi in sorted(thong_ke.keys()):
    so_luong = thong_ke[don_vi]
    ty_le = so_luong / tong_nv * 100

    print(f"{stt}./Tên đơn vị: Đơn vị {don_vi}.")
    print(f"- Số nhân viên: {so_luong}.")
    print(f"- Tỷ lệ: {ty_le:.2f}%.")
    stt += 1
```
