```python
import json
from datetime import datetime

data = []

while True:
    loai = input("Nhập loại giao dịch (tien/vang): ")
    so_luong = float(input("Nhập số lượng: "))
    don_gia = float(input("Nhập đơn giá: "))

    kq = {
        "loai": loai,
        "so_luong": so_luong,
        "don_gia": don_gia,
        "thanh_tien": so_luong * don_gia
    }

    data.append(kq)\

    tiep_tuc = input("Tiếp tục giao dịch? (y/n): ")
    if tiep_tuc.lower() != "y":
        break

lua_chon = input("Bạn có muốn ghi vào tập tin không? (1: Có, 0: Không): ")

if lua_chon == "1":
    thoi_gian = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    ten_file = f"{thoi_gian}.json"

    with open(ten_file, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print(f"Đã ghi dữ liệu vào tập tin {ten_file}")
else:
    print("Không ghi dữ liệu vào tập tin")
```
