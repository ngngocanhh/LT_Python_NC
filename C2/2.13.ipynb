```python
import json
import urllib.request

# Địa chỉ API
url = "https://jsonplaceholder.typicode.com/comments?postId=1"

# Đọc dữ liệu từ API
with urllib.request.urlopen(url) as response:
    danh_sach_bai_post = json.loads(response.read().decode())

print("Danh sách các bài post nổi bật (postId = 1)")
print("-" * 50)

# Chỉ hiển thị 3 bài post đầu tiên
for bai_post in danh_sach_bai_post[:3]:
    print("Post ID:", bai_post["postId"])
    print("ID:", bai_post["id"])
    print("Tên:", bai_post["name"])
    print("Email:", bai_post["email"])
    print("Nội dung:", bai_post["body"])
    print("-" * 50)
```
